```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import requests
import io
```

```python
url = "http://www.umich.edu/~umfandsf/other/ebooks/alice30.txt"
urlData = requests.get(url).content

file = open('data/alice.txt', 'bw')
file.write(urlData)
file.close()
```

```python
text = open('data/alice.txt').read()
```

```python
# read the mask image
mask = np.array(Image.open('data/image.png'))
```

```python
stopwords = set(STOPWORDS)
stopwords.add("said")

wc = WordCloud(background_color="white", max_words=2000, mask=mask,
               stopwords=stopwords, width=3, colormap='Reds')
```

```python
# generate word cloud
wc.generate(text)
```

```python
# store to file
wc.to_file('data/wordcloud.png')
```

```python
# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.figure()
plt.imshow(mask, cmap=plt.cm.Reds, interpolation='bilinear')
plt.axis("off")
plt.show()
```
